```python
# Read the CSV file
df = pd.read_csv("https://raw.githubusercontent.com/basamaryan/DS320FinalProject/main/master.csv")
df.tail()
```

```python
# Drop the nextdayprice and result columns
df = df.drop(['Year','Month'], axis=1)
```

```python
print(df)
```

```python
df.tail()
```

```python
# Convert the 'date' column to datetime format and set it as the index of the dataframe
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
# Define the feature and target variables
X = df.iloc[:, :-1].values
y = df.iloc[:, 4:5].values
```

```python
print(y)
```

```python
import random

random.seed(10)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# Scale the data using MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)
```

```python
# Reshape the data for LSTM
feature_length = X_train.shape[1]
X_train = X_train.reshape(X_train.shape[0], feature_length, 1)
X_test = X_test.reshape(X_test.shape[0], feature_length, 1)
```

#Building LSTM Model

```python
# Define the LSTM model
model = Sequential([
    LSTM(100, return_sequences=True, input_shape=(feature_length, 1)),
    Dropout(0.3),
    LSTM(100, return_sequences=False),
    Dropout(0.3),
    Dense(1),
])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Define the upgraded LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(feature_length, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=True),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25, activation='relu'),
    Dense(1),
])

# Compile the model
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')  # Adjust the loss function based on your problem

# Print the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=12,
    verbose=1,
)
```

```python
# Make predictions using the test set
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

##Model's Performance in test

```python
# Invert the scaling of the predictions and the actual values
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test)
```

```python
# Print the RMSE of the predictions
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

#Make Predictions

```python
# Add the 'date' column back to the dataframe
df['Date'] = df.index
```

```python
# Create a dataframe to store the predicted and actual values of closing price along with their corresponding dates
test_dates = df.loc[df.index[-len(y_test):], 'Date'].values
results = pd.DataFrame({'Date': test_dates, 'Predicted': y_pred.flatten(), 'Truth': y_test.flatten()})
```

```python
# Calculate the result column based on the predicted and actual values of closing price
results['result'] = np.where(results['Predicted'].shift(-1) > results['Predicted'], 1, 0)
```

```python
# Print the results
pd.set_option('display.max_rows', None)  # To display all the rows of the dataframe
print(results)
```

#Evaluating the Model Performance

```python
# Calculate the number of times the model correctly estimated the result column in the test set
correct_predictions = sum(results['result'][:-1] == (results['Truth'][1:].values > results['Truth'][:-1].values))

# Calculate the total number of predictions in the test set
total_predictions = len(results) - 1

# Calculate the percentage of correct predictions in the test set
accuracy = (correct_predictions / total_predictions) * 100

# Print the accuracy
print(f"The model made {correct_predictions} correct predictions out of {total_predictions} total predictions in the test set. The accuracy is {accuracy:.2f}%.")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Precision - fraction of positive predictions that are actually positive
precision = precision_score(results['Truth'][1:], results['result'][:-1])

# Recall - fraction of actual positives that are predicted positive
recall = recall_score(results['Truth'][1:], results['result'][:-1])

# F1 score - harmonic mean of precision and recall
f1 = f1_score(results['Truth'][1:], results['result'][:-1])

print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1 score: {f1:.3f}")
```
